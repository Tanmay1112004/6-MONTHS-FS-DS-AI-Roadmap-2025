```python
# Introduction to seaborn
# seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt  # Visulaization
import seaborn as sns  # distrubution visulation

%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 4) 

import warnings 
warnings.filterwarnings('ignore')


#import warnings
#warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
# Distributions: 
vis1 = sns.distplot(stats["InternetUsers"])
plt.show(vis1)
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=10) 
plt.show(vis1)
```

```python
# Box Plots
custom_colors = ["#FF9999", "#66B3FF", "#99FF99", "#FFCC99"]
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate", palette=custom_colors)
plt.show(vis2)
```

```python
vis3 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate', fit_reg = False)
#lm - linear model
plt.show(vis3)
```

```python
vis4 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate')
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
plt.show(vis5)
```

```python
vis5 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg=False, hue='IncomeGroup', height=10, aspect=1.5)
plt.show()
```
