```python
!pip install -q -U google-generativeai gradio Pillow

import gradio as gr
import google.generativeai as genai
import PIL.Image
from pathlib import Path

# Configure Gemini
genai.configure(api_key="AIzaSyDGc8fbjufsxdqus09I4lsmccpBSs-dfuI")

# Initialize models
text_model = genai.GenerativeModel("gemini-1.5-flash")
vision_model = genai.GenerativeModel('models/gemini-2.5-flash-preview-05-20')

def generate_text(prompt):
    """Generate text response from text prompt"""
    try:
        response = text_model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Error generating text: {str(e)}"

def analyze_image(img, prompt=None):
    """Analyze image with optional prompt"""
    try:
        if prompt:
            response = vision_model.generate_content([prompt, img])
        else:
            response = vision_model.generate_content(img)
        return response.text
    except Exception as e:
        return f"Error analyzing image: {str(e)}"

def combined_interface(text_prompt, image, image_prompt):
    """Handle both text and image inputs"""
    results = []

    if text_prompt:
        text_result = generate_text(text_prompt)
        results.append(("Text Response", text_result))

    if image is not None:
        img = PIL.Image.fromarray(image)
        img_result = analyze_image(img, image_prompt)
        results.append(("Image Analysis", img_result))

    if not results:
        return "Please provide either text or image input"

    return "\n\n".join([f"## {title}\n{content}" for title, content in results])

# Create Gradio interface
with gr.Blocks(title="Gemini Multi-Model Interface") as demo:
    gr.Markdown("# ðŸŒŸ Gemini Multi-Model Interface")
    gr.Markdown("Interact with Gemini's text and vision capabilities")

    with gr.Tab("Text Input"):
        text_input = gr.Textbox(label="Enter your text prompt", lines=3)
        text_output = gr.Markdown()
        text_button = gr.Button("Generate")

    with gr.Tab("Image Input"):
        image_input = gr.Image(label="Upload an image")
        image_prompt = gr.Textbox(label="Optional: Ask something about the image", lines=2)
        image_output = gr.Markdown()
        image_button = gr.Button("Analyze Image")

    with gr.Tab("Combined Input"):
        combined_text = gr.Textbox(label="Text prompt (optional)", lines=2)
        combined_image = gr.Image(label="Upload image (optional)")
        combined_img_prompt = gr.Textbox(label="Image prompt (optional)", lines=2)
        combined_output = gr.Markdown()
        combined_button = gr.Button("Process")

    # Event handlers
    text_button.click(generate_text, inputs=text_input, outputs=text_output)
    image_button.click(analyze_image, inputs=[image_input, image_prompt], outputs=image_output)
    combined_button.click(combined_interface,
                         inputs=[combined_text, combined_image, combined_img_prompt],
                         outputs=combined_output)

# Launch the app
demo.launch(debug=True)
```
